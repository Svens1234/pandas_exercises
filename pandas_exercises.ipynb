```python
#Create a table for counts for each day of the week that people arrived. (E.g. 5000 arrivals were on a Monday, 3000 were on a Tuesday, etc..)
```

```python
import numpy as np
```

```python
def convert(day,month,year):
    return f'{day}-{month}-{year}'
```

```python
hotels['date']  = np.vectorize(convert)(hotels['arrival_date_day_of_month'],
                                        hotels['arrival_date_month'],
                                        hotels['arrival_date_year'])
```

```python
hotels['date'] = pd.to_datetime(hotels['date'])
```
